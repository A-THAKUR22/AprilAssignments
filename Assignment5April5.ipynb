Applying Decision tree classification on the Iris dataset in the sklearn library to predict the species of the flower based on the features given.

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import svm
import seaborn as sns
```

```python
iris_data=load_iris()
iris=pd.DataFrame(data=iris_data.data,columns=iris_data.feature_names)
iris['Species']=iris_data.target
```

```python
iris
```

```python
sns.histplot(data=iris,x='sepal length (cm)',hue='Species',kde=True)
```

```python
sns.histplot(data=iris,x='sepal width (cm)',hue='Species',kde=True)
```

```python
sns.histplot(data=iris,x='petal length (cm)',hue='Species',kde=True)
```

```python
sns.histplot(data=iris,x='petal width (cm)',hue='Species',kde=True)
```

```python
X=iris.drop(columns='Species')
y=iris['Species']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.8,random_state=42)
treeclf=DecisionTreeClassifier()
treeclf.fit(X_train,y_train)
```

```python
plt.figure(figsize=(15,10))
plot_tree(treeclf,filled=True)
```

```python
y_pred=treeclf.predict(X_test)
print(f'Predictions of species of the flower : {y_pred}')
```

```python
Accuracy=accuracy_score(y_test,y_pred)
print(f'The accuracy score of our model : {Accuracy}')
```

```python
ypred2=treeclf.predict(X_train)
acc2=accuracy_score(y_train,ypred2)
acc2
```

```python
svclf=svm.SVC(kernel='linear')
svclf.fit(X_train,y_train)
```

```python
svpred=svclf.predict(X_test)
svpred
```

```python
acc3=accuracy_score(y_test,svpred)
acc3
```

```python
svpred2=svclf.predict(X_train)
acc4=accuracy_score(y_train,svpred2)
acc4
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randclf=RandomForestClassifier()
randclf.fit(X_train,y_train)
```

```python
pred_for=randclf.predict(X_test)
pred_for
```

```python
for_acc=accuracy_score(y_test,pred_for)
for_acc
```

```python
for_pred2=randclf.predict(X_train)
acc5=accuracy_score(y_train,for_pred2)
acc5
```

```python
from mlxtend.evaluate import bias_variance_decomp
```

```python
mse,bias,varaince=bias_variance_decomp(
    estimator=treeclf,
    X_train=X_train.values,
    y_train=y_train.values,
    X_test=X_test.values,
    y_test=y_test.values,
    random_seed=42,
    loss='mse'
    )
print(mse)
print(bias)
print(varaince)
```

```python
mse,bias,varaince=bias_variance_decomp(
    estimator=svclf,
    X_train=X_train.values,
    y_train=y_train.values,
    X_test=X_test.values,
    y_test=y_test.values,
    random_seed=42,
    loss='mse'
    )
print(mse)
print(bias)
print(varaince)
```
